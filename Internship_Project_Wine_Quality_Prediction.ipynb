```python
import pandas as pd
```

```python
WineQuality = pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/WhiteWineQuality.csv',sep = ';')
```

```python
y = WineQuality['quality']
```

```python
X = WineQuality[['fixed acidity',"volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]]
```

```python
X
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2529)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```
